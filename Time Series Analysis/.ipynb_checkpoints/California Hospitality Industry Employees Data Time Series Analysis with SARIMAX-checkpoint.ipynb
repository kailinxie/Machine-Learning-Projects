```python
import pandas as pd
import numpy as np
%matplotlib inline

# Load specific forecasting tools
from statsmodels.tsa.statespace.sarimax import SARIMAX

from statsmodels.tsa.seasonal import seasonal_decompose      # for ETS Plots
from pmdarima import auto_arima                              # for determining ARIMA orders

# Load specific evaluation tools
from sklearn.metrics import mean_squared_error
from statsmodels.tools.eval_measures import rmse

# Ignore harmless warnings
import warnings
warnings.filterwarnings("ignore")

# Load datasets, set date as index column
df = pd.read_csv('HospitalityEmployees.csv', index_col='Date', parse_dates = True)
df.index.freq = 'MS'
print(len(df))
df
```

So <tt>df</tt> has 348 records and one column. The data represents the number of employees in thousands of persons as monthly averages from January, 1990 to December 2018.

### 1. Plot the source data

```python
df.plot(figsize = (12, 5), legend = False, ylabel = 'Thousands of Persons', title = 'California Hospitality Industry Employees');
```

### 2. Run an ETS Decomposition
Use an 'additive' model.

```python
seasonal_decompose(df['Employees'], model = 'add').plot();
```

### 3. Run <tt>pmdarima.auto_arima</tt> to obtain recommended orders

```python
auto_arima(df['Employees'], seasonal = True, m = 12).summary()
```

### 4. Split the data into train/test sets

```python
train = df.iloc[:-12]
test = df.iloc[-12:]
```

### 5. Fit a SARIMA(1,1,2)(1,0,1,12) model to the training set

```python
model = SARIMAX(train['Employees'], order = (1, 1, 2), seasonal_order = (1, 0, 1, 12))
results = model.fit()
results.summary()
```

### 6. Obtain predicted values

```python
predictions = results.predict(start = len(train), end = len(df) - 1, dynamic = False, typ = 'levels')
```

### 7. Plot predictions against known values

```python
ax = test['Employees'].plot(legend = True, figsize = (12, 6), ylabel = 'Thousands of Persons', title = 'California Hospitality Industry Employees')
predictions.plot(legend = True);
```

### 8. Evaluate the Model using MSE and RMSE

```python
print(f"MSE: {mean_squared_error(predictions, test['Employees'])}")
print(f"RMSE: {rmse(predictions, test['Employees'])}")
```

```python
test['Employees'].mean()
```

### 9. Retrain the model on the full data and forecast one year into the future

```python
model = SARIMAX(df['Employees'], order = (1, 1, 2), seasonal_order = (1, 0, 1, 12))
results = model.fit()
results.summary()
forecast = results.predict(start = len(df), end = len(df) + 11, typ = 'levels')
```

### 10. Plot the forecasted values alongside the original data

```python
ax = df['Employees'].plot(legend = True, figsize = (12, 6), title = 'California Hospitality Industry Employees', ylabel = 'Thousands of Persons')
forecast.plot(legend = True);
```
